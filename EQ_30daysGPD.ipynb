```python
import pandas as pd 
import matplotlib.pyplot as plt 
import geopandas as gpd
```

### Load in csv file to a Pandas Dataframe and inspect data

```python
df = pd.read_csv('2.5_month.csv', usecols = ['latitude', 'longitude', 'mag', 
                                             'updated', 'time', 'place'], 
                                             parse_dates=['time'])

df.head()
```

### Plot data on scatter plot, latitude on the y-axis and longitude on the x-axis

```python
df.plot(x="longitude", y="latitude", kind="scatter")
```

### Create a GeoPandas map object

```python
countries = gpd.read_file(
               gpd.datasets.get_path("naturalearth_lowres"))
countries.plot(color="lightgrey")
```

### Putting it all together

```python
#intialize an axis 
fig, ax = plt.subplots(figsize=(8,6)) 

#plot map on axis
countries = gpd.read_file(
    gpd.datasets.get_path("naturalearth_lowres")) 
countries.plot(color="lightgrey", ax=ax)

#plot points 
df.plot(x="longitude", y="latitude", kind="scatter", 
        title="Global Earthquakes For The Past 30 Days", ax=ax, legend=True) 

#add grid
ax.grid(visible=True, alpha=0.5)
plt.show()
```
